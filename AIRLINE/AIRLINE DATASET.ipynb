### MODELING

```python
x = df['Airport Name'].values.reshape(-1,1)
y = df['Flight Status'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train, y_train)
```

```python
#nilai m dan b
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
#nilai score accurasy 
lin_reg.score(x_test, y_test)
```

##### *modeling ini mendapatkan tingkat keakurasian 1.0 atau 100%

```python
y_prediksi = lin_reg.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediksi, c='red',)
plt.xlabel('Airport Name')
plt.ylabel('Flight Status')
plt.title('Plot Airport Name & Flight Status')
```
